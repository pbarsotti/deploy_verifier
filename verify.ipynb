### Verify pools and deploy version

```python
import pandas as pd

df = pd.read_csv('pools/pools.csv')
```

```python
df.head()

EXPECTED_JOBS_VERSION='0.60.26'
EXPECTED_API_VERSION='0.56.38'
EXPECTED_SEARCH_VERSION=''
EXPECTED_MAILING_VERSION='0.1.14'
EXPECTED_TRACKING_VERSION='0.1.35'

```

```python
import requests
import json

get_instancer_per_pool='http://api.melicloud.com/compute/pools/{0}/instances'
get_instance_details='http://api.melicloud.com/monitoring/instances/{0}'

i=0
while i < df.shape[0]:
    pool=df.iloc[i][0]
    response = requests.get(get_instancer_per_pool.format(pool))
    instances=response.json()
    expected_version=defineApplication(pool)
    if expected_version != "":
        j=0
        had_issues =False
        print(f"\33[0m Pool: {pool} \33[0m")
        while j < len(instances):
            instance=instances[j]
            response = requests.get(get_instance_details.format(instance))
            details=response.json()
            if details['version'] != expected_version:
                print(f"\33[33m           Verify pool {pool} for instance {instance}\33[33m")
                had_issues=True
            j+=1
        if had_issues == False:
            print("           \033[32m Ok \033[32m")
    else:
        print(f"\33[36m Skipping pool: {pool}\33[36m")

    i+=1

print('End')





```

```python
def defineApplication(pool):
    if "shipping-api" in pool:
        return EXPECTED_API_VERSION
    if "shipping-jobs" in pool:
        return EXPECTED_JOBS_VERSION
    if "shipping-search" in pool:
        return EXPECTED_SEARCH_VERSION
    if "shipping-tracking" in pool:
        return EXPECTED_TRACKING_VERSION
    if "shipping-mailing" in pool:
        return EXPECTED_MAILING_VERSION
    return ""
        
```
